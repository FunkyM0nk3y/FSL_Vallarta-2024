## Ciclos

En esta sección solo vamos a ver el caso del ciclo *for* aunque en Python existen otras tantas opciones para iterar y hacer ciclos.

```
listaDeNombres = ["Juan", "Elly", "Pedro", "María"]

print("La lista tiene " + str(len(listaDeNombres)) + " nombres")
print("\n")
```

```
for turno in range(len(listaDeNombres)):
  print("Ahora es el turno de " + listaDeNombres[turno])
  print("\tPasa a saltar la cuerda\n")
```

El principio de cómo funciona el ciclo for es muy simple.  Se utiliza una variable (en este caso turno) como transporte de los valores al bloque interno de código.  El ciclo *for* va iterando en orden por si solo por cada elemento que recibe como argumento hasta pasar por todos.

Adicional a esto podrás observar que es muy fácil utilizar carácteres de control (*\n, \t*) interpolados en cadenas.

La siguiente función va a servir como un validador de tipo de dato.  Se podrían generar funciones para validar cada tipo con el que vamos a trabajar, o bien, extender su funcionalidad para recibir el tipo a validar.

```
def validaEsTexto(texto):
  if isinstance(texto, str):
    return texto
  else:
    return None

entero = 3
flotante = 3.14
cadena = 'pi'
```

```
type(validaEsTexto(entero))
```

```
type(validaEsTexto(flotante))
```

```
type(validaEsTexto(cadena))
```

La función *validaEsTexto()* recibe un argumento que es accesible a partir de la variable *texto*, a esta se le evalua el tipo de dato y se devuelve un valor de acuerdo a nuestras necesidades.

```
def buscaPokemon(nombrePokemon, listaPokemon):
  if not validaEsTexto(nombrePokemon):
    print("nombrePokemon debe ser texto.")
    return None

  for counter in range(len(listaPokemon)):
    if(listaPokemon[counter]['name'] == nombrePokemon):
      return listaPokemon[counter]
  return None
```

En esta última función estamos utilizando a la que definimos previamente (*validaEsTexto*) para validar que el tipo de dato que estamos recibiendo tiene sentido de acuerdo a la necesidad que tenemos (recibir nombres de pokemones).

En caso de cumplirse la condición barremos la lista de pokemones recibida y si el nombre del pokemon está en la lista devolvemos los datos que necesitamos.

Ahora veamos cómo utilizarla.

```
listaPokemon = requests.get("https://pokeapi.co/api/v2/pokemon")
if listaPokemon.status_code == 200:
  listaPokemon = listaPokemon.json()['results']

listaPokemon
```

Se ve más complicado de lo que en realidad es el código anterior.

Veamos, primero solicitamos al ***API*** de Pokemon que nos dé los datos que tiene disponibles.  Para esto utilizamos la biblioteca ***requests***, cuando hacemos una solicitud exitosa a un API, regularmente tendremos de regreso un status ***200***, esto es lo primero que validamos.  Lo siguiente que hacemos es formatear un poco los datos para trabajar con ellos.

### Filtrando por valores

Comencemos limpiando los valores no numéricos que tenemos en la columna "mass".


Primero veamos cuáles valores podrían meternos en problemas.

Para esto vamos a utilizar expresiones regulares.


```
# [ ] Se utiliza para decir que puede ser cualquier opción de las que están ahí dentro.
# a-z Se usa para definir el rango de "letras" de la 'a' a la 'z'.
# A-Z Se usa para definir el rango de "letras" de la 'A' a la 'Z'.
# El repetidor '+' quiere decir una o más veces el patrón.
regex = '[a-zA-Z]+'
dataframe[dataframe.mass.str.contains(regex)]
```

regex-cartoon.jpg

Las expresiones regulares (**regex**), pueden parecer un lenguaje oscuro olvidado, pero...  son muy poderosas para encontrar patrones.

Usando filtros con expresiones regulares podemos encontrar todos los valores "extraños" que se tengan para la columna "mass" y eliminarlos.  En el siguiente código vamos a buscar todos lo valores que tienen números incluyendo decimales y solo dejar estos en el DataFrame con el que trabajaremos.

```
# \d se utiliza para hacer referencia a un valor numérico [0-9].
# Los anclajes '^' y '$' se usan para especificar el principio y final.
# El repetidor '*' quiere decir cero o más veces el patrón.
# El repetidor '?' quiere decir que puede o no aperecer el patrón anterior (una o cero veces).
regex = '^\d+\.?\d*$'
df_nostring = dataframe[dataframe.mass.str.contains(regex)]

convert_dictionary= {
    'fall':str,
    'mass':float,
    'name':str,
    'reclat':float,
    'reclong':float,
    'recclass':str,
    'year':str
}
df_converted = df_nostring.astype(convert_dictionary)
df_converted.dtypes
```

## Fase 3: Uso de los datos **limpios**

```
df_large = df_large.sort_values('recclass', ascending=True)

plt.rcParams['figure.figsize'] = [30, 20]
plt.grid(True, color='lightgray', linestyle='-.', linewidth=1)
plt.gca().patch.set_facecolor('1')

font_title = {'family': 'serif',
              'color':  'black',
              'weight': 'bold',
              'size': 13,
             }

plt.scatter(df_large['mass'], df_large['recclass'])
plt.title('Tamaño de masa por clasificación', fontdict=font_title)
plt.xlabel('Masa')
plt.ylabel('Clasificación')
plt.subplots_adjust(bottom=0.5)
plt.show()
```

```
fig, axs = plt.subplots(1, 2, figsize=(30, 15), sharey=True)
plt.gca().patch.set_facecolor('1')

axs[0].bar(df_large['recclass'], df_large['year'])
axs[1].scatter(df_large['recclass'], df_large['year'])
axs[0].grid(True, color='lightgray', linestyle='dotted', linewidth=1)
axs[1].grid(True, color='lightgray', linestyle='dotted', linewidth=1)
fig.suptitle('Categorías vs Año', y=0.9, fontsize=13)

fig.text(0.5, 0.02, 'Categorías', ha='center', va='center')
fig.text(0.1, 0.5, 'Años', ha='center', va='center', rotation='vertical')

for ax in fig.axes:
  ax.tick_params(axis='x', labelrotation=90)
```

Todo incida que la representación por barras para este problema no es la mejor opción.  Se pierde cuáles han sido registrados en más de un año.

Para que tenga más sentido la gráfica vamos a ordenar las observaciones por los años antes de graficar los datos y solo vamos a trabajar el scatter plot.

Vamos a representar los puntos por el valor de "mass".  ¿Qué puedes observar?
Remueve **alpha=alpha_val** y observa qué pasa.  Juega con distintos valores para **alpha_val**.

```
alpha_val = 0.3
df_large = df_large.sort_values('year', ascending=True)

plt.rcParams['figure.figsize'] = [30, 15]
plt.grid(True, color='gray', linestyle='-.', linewidth=0.5)
plt.gca().patch.set_facecolor('1')

font_t = {'family': 'serif',
        'color':  'black',
        'weight': 'bold',
        'size': 13,
        }
font_st = {'family': 'serif',
        'color':  'gray',
        'weight': 'bold',
        'size': 11,
        }

plt.scatter(df_large['recclass'], df_large['year'], s=df_large['mass'], alpha=alpha_val)
plt.title('Categorías por año', fontdict=font_t)
plt.xlabel('Categoría', fontdict=font_st)
plt.ylabel('Año', fontdict=font_st)
plt.show()
```

Vamos a normalizar "mass" para poder apreciar en la gráfica los tamaños relativos.

Una vez normalizados los datos algunos puntos pueden desaparecer.  Para poder representarlos vamos a ponerles un valor "artificial" y denotarlos con otro color.

```
df_large['norm_mass'] = ((df_large['mass'] - df_large['mass'].min()) /
                         (df_large['mass'].max() - df_large['mass'].min()))*50000
df_large['on_scale'] = np.where(df_large['norm_mass'] != 0, 'navy', 'red')
df_large.loc[df_large['norm_mass'] == 0, 'norm_mass'] = (df_large['norm_mass'].sort_values(ascending=True).unique())[1]
```

Veamos cómo lucen ahora los puntos normalizados y aprovechemos para ponerles nombre.
